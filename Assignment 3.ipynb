### 6.6 (c)

```python
T = np.linspace(0.0001, 10, 1000)
```

```python
def Cv(T):
    C = T**(-2) * exp(-1/T) * (1 + exp(-1/T))
    return C
```

```python
plt.figure(figsize=(10, 4))

plt.plot(T, Cv(T))
plt.xlabel(r'Temperature ($T$)')
plt.ylabel(r'$\frac{C_V(T) k}{N (\varepsilon_2 - \varepsilon_1)^2} $', labelpad=30, rotation=0, fontsize=14)
plt.title(r'Relationship between heat capacity ($C_V(T)$) and temperature ($T$)', fontsize=14)
plt.show()
```
